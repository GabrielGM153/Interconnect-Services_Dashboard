# Interconnect (Proyecto Final parte I)

Al operador de telecomunicaciones Interconnect le gustaría poder pronosticar su tasa de cancelación de clientes. Si se descubre que un usuario o usuaria planea irse, se le ofrecerán códigos promocionales y opciones de planes especiales. El equipo de marketing de Interconnect ha recopilado algunos de los datos personales de sus clientes, incluyendo información sobre sus planes y contratos.

### Servicios de Interconnect

Interconnect proporciona principalmente dos tipos de servicios:

1. Comunicación por teléfono fijo. El teléfono se puede conectar a varias líneas de manera simultánea.

2. Internet. La red se puede configurar a través de una línea telefónica (DSL, *línea de abonado digital*) o a través de un cable de fibra óptica.

Algunos otros servicios que ofrece la empresa incluyen:

- Seguridad en Internet: software antivirus (*ProtecciónDeDispositivo*) y un bloqueador de sitios web maliciosos (*SeguridadEnLínea*).

- Una línea de soporte técnico (*SoporteTécnico*).

- Almacenamiento de archivos en la nube y backup de datos (*BackupOnline*).

- Streaming de TV (*StreamingTV*) y directorio de películas (*StreamingPelículas*)

La clientela puede elegir entre un pago mensual o firmar un contrato de 1 o 2 años. Puede utilizar varios métodos de pago y recibir una factura electrónica después de una transacción.

### Plan de trabajo

Al leer la problemática que enfrenta Interconnect, se evalúa que dicha empresa se enfrenta a una problemática de retención de clientes. Así mismo, y al no tener información acerca del mercado y su estructura, el análisis se acota a los datos internos de la empresa; mismos que corresponden a datos de clientes, paquetes de telefonía, internet y los contratos (tanto vigentes, como vencidos).

Siguiendo la naturaleza de los datos, y en aras de resolver la problemática que se enuncia, se propone el siguiente plan de trabajo:

1) Descripción de datos: 

- Orientar el análisis en términos de qué género es el principal consumidor del servicio, este componente se propone analizar quienes son los más receptivos al trabajo de márketing actual, y a partir de ahí hacer ajustes en dicha área.

- Separar la demanda de tipo de servicio de internet, sea de DSL o fibra óptica.

- Analizar la demanda de seguridad online, servicio de streaming, almacenamiento en la nube, así como servicio de soporte técnico. El analizar esta área de la oferta de la empresa ayudará, y al tener datos sólidos del comportamiento de la demanda de servicios, permitirá a la empresa tomar mejores decisiones en términos de la oferta de dichos servicios, así como la propia asignación de recursos en el área que se requieran.

- También se ofrecerá un análisis dewtallado del comportamiento de la demanda de servicios principales: Telefonía e internet.

2) Para medir la probabilidad de cancelación:

- Se analizará dicha probabilidad con ayuda de una regresión logística, este análisis dotará a la compañá de una capacidad de replanteamiento interno en cuanto a la oferta del servicio. Además, este análisis busca ser concreto e integral con la narrativa de datos que se describe en el punto 1) 

Por otra parte, la probabilidad de cancelación ayuda a la compañía a hacer inferencias con respecto a la estructura del mercado y a la competencia a que se enfrenta. 

- En aras de reforzar ma medición, se ofrece el entrenamiento de un modelo de bosque aleatorio.

**Se usará de base el modelo que obtenga las mejores métricas:**

a) Accuracy.

b) Precision.

c) Recall.

d) F1 Score.

e) Curva Roc.

f) Matriz de confusión.


3) Posteriormente, habiendo estudiado minuciosamente la narrativa de los datos, los resultados obtenidos bajo los modelos descritos, así como las métricas correspondientes, se elaborará un reporte completo y explicativo donde se ofrecerá el posible camino a seguir.
